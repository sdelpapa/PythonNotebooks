Blackjack - Regras:
    Jogador recebe primeira carta aberta
    Mesa primeira carta aberta
    Joagador recebe segunda carta aberta
    Mesa segunda carta fechada
    Jogador decide "continuar" ou "ficar"
        Se estourar 21, mesa ganha
    Qdo jogador ficar, mesa abre a segunda carta
        Se valor abaixo de 17, puxa cartas, até atingir valor >=17 ou estourar
        Se mesa estourar, jogador ganha
    Quem tiver o jogo maior ganha

```python
import random
from IPython.display import clear_output

class Baralho(object):
    def __init__(self):
        
        self.carta_atual = 0
        self.ordem_das_cartas =[]
        
        sinal = ["A"]
        sinal += [str(i) for i in range(2,11)]
        sinal += ["J", "Q", "K"]

        #Ordem naipe = paus, copas, espadas, ouros
        naipe = ["PAUS", "COPAS", "ESPADAS", "OUROS"]

        valor = [i for i in range(1,11)]
        valor += [10, 10, 10]

        #Matriz cartas = indice, naipe, sinal, valor
        self.cartas = []
        i = 0
        for x in range(0,4):
                for j in range(0,13):
                    self.cartas.append([i, naipe[x], sinal[j], valor[j]])
                    i += 1        
        
        
    def Embaralhar(self):            
        self.ordem_das_cartas = random.sample(range(0, 13*4), 13*4)
    
        
    def ProximaCarta(self):
        self.carta_atual += 1
        if self.carta_atual > 52:
            return "FIM DO BARALHO"
        
        return self.cartas[self.ordem_das_cartas[self.carta_atual]]
    
    
class Jogador(object):
    def __init__(self):
        self.cartas = []
        self.pontos = 0

    def ContinuaJogada(self):
        marker = ''
        while not (marker == 'M' or marker == 'F'):
            marker = input('JOGADOR quer Mais cartas ou Fica? Insira "M" ou "F".').upper()[0]
            
        return marker

    def Replay(self):
        return input('Quer jogar de novo? "SIM" ou "NAO"').lower().startswith('s')    


class Mesa(object):
    def __init__(self):
        self.cartas = []
        self.pontos = 0

    def ContinuaJogada(self):
        marker = ''
        if self.pontos < 17:
            #while not (marker == 'M' or marker == 'F'):
            #    marker = input('MESA quer Mais cartas ou Fica? Insira "M" ou "F".').upper()[0]
            marker = "M"
        else:
            marker = "F"
            
        return marker
    
    
class Jogo(object):
    def __init__(self):
        self.aberto = False
        self.fim = False
        
        self.Jogador = Jogador()
        self.Baralho = Baralho()
        self.Mesa = Mesa()

    def Inicia(self):        
        self.__init__()
        self.Embaralhar()
        self.DarCartas()

    def Embaralhar(self):
        self.Baralho.Embaralhar()
        
    def ChecaGanhador(self):
        self.msg = ""
        if self.Jogador.pontos > 21 and self.Mesa.pontos > 21:
            self.fim = True
            self.aberto = True
            self.msg = "Ninguém ganhou!!!!"
        elif self.Jogador.pontos > 21:
            self.fim = True
            self.aberto = True
            self.msg = "Mesa ganhou!!!!"
        elif self.Mesa.pontos > 21:
            self.fim = True
            self.aberto = True
            self.msg = "Jogador ganhou!!!!"
        elif self.fim:
            if self.Jogador.pontos == self.Mesa.pontos:
                self.msg = "Ninguém ganhou!!!!"
            elif self.Jogador.pontos > self.Mesa.pontos:
                self.msg = "Jogador ganhou!!!!"
            else:
                self.aberto = True
                self.msg = "Mesa ganhou!!!!"
            
        return self.fim
        
    def CalculaPontos(self):
        tem_as = False
        tem_dez = False
            
        self.Mesa.pontos = 0
        for i in self.Mesa.cartas:
            if i[3] == 1: 
                tem_as = True
            if i[3] == 10: 
                tem_dez = True

            self.Mesa.pontos += i[3] 

        if tem_as and tem_dez and ((self.Mesa.pontos + 10) <= 21):
            self.Mesa.pontos += 10
        
            
        tem_as = False
        tem_dez = False
            
        self.Jogador.pontos = 0
        for i in self.Jogador.cartas:
            if i[3] == 1: 
                tem_as = True
            if i[3] == 10: 
                tem_dez = True

            self.Jogador.pontos += i[3] 

        if tem_as and tem_dez and ((self.Jogador.pontos + 10) <= 21):
            self.Jogador.pontos += 10
            
  
    def DarCartas(self):
        self.ProximaCartaJogador()
        self.ProximaCartaMesa()
        self.ProximaCartaJogador()
        self.ProximaCartaMesa()
        
        self.aberto = False
        
    def ProximaCartaJogador(self):
        self.Jogador.cartas += [self.Baralho.ProximaCarta()]
        self.CalculaPontos()
        self.ChecaGanhador()
        self.Exibe()

    def ProximaCartaMesa(self):
        self.Mesa.cartas += [self.Baralho.ProximaCarta()]
        self.CalculaPontos()
        self.ChecaGanhador()
        self.Exibe()
    
    def AbreMesa(self):
        self.aberto = True
        self.ChecaGanhador()
        self.Exibe()

    def Fim(self):
        self.fim = True
        self.ChecaGanhador()
        self.Exibe()
        
    def Exibe(self):

        clear_output()
        
        if self.aberto:
            print("MESA / Pontos = ", self.Mesa.pontos)
            for i in self.Mesa.cartas:
                print("------", i[2] + "-" + i[1])
        else:
            if len(self.Mesa.cartas) > 0:
                print("MESA / Pontos = X")
                print("------", self.Mesa.cartas[0][2] + "-" + self.Mesa.cartas[0][1])
                print("------", "XXXXXXXXX")
            
        print("JOGADOR / Pontos = ", self.Jogador.pontos)
        for i in self.Jogador.cartas:
            print("------", i[2] + "-" + i[1])
        
        if self.fim:
            print("###############", self.msg, "###############")

    
```

```python
print("Bem-vindo ao jogo de BlackJack!!!")

jogada = ""

while True:

    j = Jogo()
    j.Inicia()

    game_on = True

    while game_on:
        if not j.fim:
            if not j.aberto:
                jogada = j.Jogador.ContinuaJogada()
                if jogada == "F":
                    j.AbreMesa()
                else:
                    j.fim = j.ProximaCartaJogador()
            else:
                jogada = j.Mesa.ContinuaJogada()
                
                if jogada == "F":
                    game_on = False 
                    j.Fim()
                else:
                    j.fim = j.ProximaCartaMesa()
        else:
            game_on = False
            
            
    if not j.Jogador.Replay():
        break
        
```
