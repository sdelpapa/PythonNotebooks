```python
from IPython.display import clear_output

jogador1 = ""    
jogador2 = ""
jogadas = []
marker =  ""
game_on = False
turn = ""

def LimpaJogo():
    global jogadas
    jogadas.clear
    for i in range(3):
        jogadas += str(i)
        jogadas[i] = [" ", " ", " "]



def RecebeJogada(jogador, linha_jogada, coluna_jogada):

    global jogadas

    if linha_jogada < 1 or coluna_jogada < 1 or  linha_jogada > 3 or coluna_jogada > 3:
        print("Os valores das linhas e colunas devem ser entre 1 e 3")
        return False

    #print(jogador, linha_jogada, coluna_jogada)
    
    linha_jogada += -1  
    coluna_jogada += -1

    if jogadas[linha_jogada][coluna_jogada] != " ":
        print("Jogador " + turn + ", a célula " + str(linha_jogada+1) + "x" + str(coluna_jogada+1) + " já contem jogada.")
        return False

    jogadas[linha_jogada][coluna_jogada] = jogador
    ImprimeTabuleiro()
    
    return True
    
def ChecaGanhador():
    msg_ganhou = "Jogador " + turn + " ganhou!!!"
    ganhou_linha = 0
    ganhou_coluna = 0
    ganhou_diagonal1 = 0
    ganhou_diagonal2 = 0
    celulas_preenchidas = 0
    
    for i in range(3):
        for l in range(3):
            if jogadas[i][l] != " ":
                celulas_preenchidas += 1
                
            if jogadas[i][l] == jogador:
                ganhou_linha += 1
                #print("linha", i+1, l+1, ganhou_linha, jogador, jogadas)
                
            if jogadas[l][i] == jogador:                
                ganhou_coluna += 1
                #print("coluna", l+1, i+1,  ganhou_coluna, jogador, jogadas)

            if (jogadas[i][l] == jogador) and (i == l):                
                ganhou_diagonal1 += 1
                #print("diagonal1", i+1, l+1,  ganhou_diagonal1, jogador, jogadas)

            if (jogadas[i][l] == jogador) and ((i == 1 and l == 1) or ((i == (l - 2)) or (l == (i - 2)))):                
                ganhou_diagonal2 += 1
                #print("diagonal2", i+1, l+1,  ganhou_diagonal2, jogador, jogadas)
                                
            if ganhou_linha == 3 or ganhou_coluna == 3: 
                game_on = False
                print(msg_ganhou)
                return  True

            
        if ganhou_diagonal1 == 3 or ganhou_diagonal2 == 3:
            game_on = False
            print(msg_ganhou)
            return  True
                
        if celulas_preenchidas == 9:
            game_on = False
            print("Jogo empatado!")
            return  True


        ganhou_linha = 0
        ganhou_coluna = 0
        
    return  False
        
def ImprimeTabuleiro():
    # Limpa o output
    clear_output()
    # Printa o tabuleiro
    print("  "+jogadas[0][0]+" | "+jogadas[0][1]+" | "+jogadas[0][2]+" ")
    print("------------")
    print("  "+jogadas[1][0]+" | "+jogadas[1][1]+" | "+jogadas[1][2]+" ")
    print("------------")
    print("  "+jogadas[2][0]+" | "+jogadas[2][1]+" | "+jogadas[2][2]+" ")
              

def player_input():
    
    marker = ''
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1: Você quer ser X ou O?').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
        
def player_choice():
    
    linha = 0
    while not (linha > 0 and linha < 4):
        linha = int(input('Jogador ' + str(turn) + ', escolha a linha (de 1 a 3): '))
    
    coluna = 0
    while not (coluna > 0 and coluna < 4):
        coluna = int(input('Jogador ' + str(turn) + ', escolha a coluna (de 1 a 3): '))
    
    return int(linha), int(coluna)
    
    
def replay():
    return input('Quer jogar de novo? "SIM" ou "NAO"').lower().startswith('s')
```

```python
print('Bem vindo ao jogo da velha')

import random

while True:
    LimpaJogo()
        
    jogador1, jogador2 = player_input()
    turn = str(random.randrange(1,3))
    #if turn == 2: turn = 'ROBÔ'
    print('Jogador ' + turn + ' começa!')
    game_on = True

    while game_on:
        if turn == '1':     
            jogador = jogador1
            linha, coluna = player_choice()
        else:
            jogador = jogador2
            linha = int(random.randrange(1,4))
            coluna = int(random.randrange(1,4))
            
        
        jogada_sucess = RecebeJogada(jogador, linha, coluna)
        ImprimeTabuleiro
        game_on = not(ChecaGanhador())

        if jogada_sucess:
            if turn == '1':     
                turn = 'ROBÔ'
            else:
                turn = '1'
        
    if not replay():
        break
```
